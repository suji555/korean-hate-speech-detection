```python
from torchtext.legacy.data import Field # torchtext.data 임포트
from konlpy.tag import Mecab
```

```python
# Mecab을 토크나이저로 사용
tokenizer = Mecab()
```

```python
# 필드 정의
TEXT = Field(sequential=True,
             use_vocab=True,
             tokenize=tokenizer.morphs, # 토크나이저로는 Mecab 사용.
             init_token = '<sos>',
             eos_token = '<eos>',
             lower=True,
             batch_first=True,
             fix_length=20)

LABEL = Field(sequential=False,
              use_vocab=False,
              is_target=True)
```

```python
from torchtext.legacy.data import TabularDataset
```

```python
train_data, valid_data = TabularDataset.splits(path='/content/drive/MyDrive/data/korean-hate-speech-detection/', 
                            train='train.csv',
                            test='valid.csv',
                            format='csv', 
                            fields=[('comments', TEXT), ('label', LABEL)],
                            skip_header=True)
```

```python
print('훈련 샘플의 개수 : {}'.format(len(train_data)))
print('테스트 샘플의 개수 : {}'.format(len(valid_data)))
```

```python
batch_size = 8
train_loader = Iterator(dataset=train_data, 
                        batch_size = batch_size,
                        device=None)
valid_loader = Iterator(dataset=valid_data, 
                        batch_size = batch_size,
                        device=None)
```

```python
n_vocab = len(TEXT.vocab)
n_trg = 3
pad_idx = 1
bos_idx = 2
eos_idx = 3
unk_idx = 0
```

```python
USE_CUDA = torch.cuda.is_available()
DEVICE = torch.device("cuda" if USE_CUDA else "cpu")
print("cpu와 cuda 중 다음 기기로 학습함:", DEVICE)
```

```python
from models import Transformer

model = Transformer(
            n_vocab, n_trg, pad_idx,
            d_word_vec=512, d_model=512, d_inner=2048,
            n_layers=12, n_head=8, d_k=64, d_v=64, dropout=0.1, n_position=200,
            emb_prj_weight_sharing=True,
            scale_emb_or_prj='prj').to(DEVICE)
```

```python
import torch.optim as optim

lr = 1e-2

optimizer = optim.Adam(model.parameters(), lr=lr, betas=(0.9, 0.999), eps=1e-08, weight_decay=1e-5)
```

```python
def train(model, optimizer, train_loader):
    model.train()
    corrects, total_loss = 0, 0
    for b, batch in enumerate(train_loader):
        x, y = batch.comments.to(DEVICE), batch.label.to(DEVICE)
        optimizer.zero_grad()
        logit = model(x)
        loss = F.cross_entropy(logit, y)
        total_loss += loss.item()
        corrects += (logit.argmax(1).view(y.size()).data == y.data).sum()
        loss.backward()
        optimizer.step()
    size = len(train_loader.dataset)
    avg_loss = total_loss / size
    avg_accuracy = 100.0 * corrects / size
    return avg_loss, avg_accuracy
```

```python
def evaluate(model, valid_loader):
    """evaluate model"""
    model.eval()
    corrects, total_loss = 0, 0
    for batch in valid_loader:
        x, y = batch.comments.to(DEVICE), batch.label.to(DEVICE)
        logit = model(x)
        loss = F.cross_entropy(logit, y, reduction='sum')
        total_loss += loss.item()
        corrects += (logit.argmax(1).view(y.size()).data == y.data).sum()
    size = len(valid_loader.dataset)
    avg_loss = total_loss / size
    avg_accuracy = 100.0 * corrects / size
    return avg_loss, avg_accuracy
```

```python
import os
```

```python
EPOCHS = 20
best_val_loss = None
for e in range(1, EPOCHS+1):
    train_loss, train_accuracy = train(model, optimizer, train_loader)
    val_loss, val_accuracy = evaluate(model, valid_loader)

    print("[Epoch: %d] train loss : %5.2f | train accuracy : %5.2f | val loss : %5.2f | val accuracy : %5.2f" % (e, train_loss, train_accuracy, val_loss, val_accuracy))

    # 검증 오차가 가장 적은 최적의 모델을 저장
    if not best_val_loss or val_loss < best_val_loss:
        if not os.path.isdir("/content/drive/MyDrive/snapshot"):
            os.makedirs("/content/drive/MyDrive/snapshot")
        torch.save(model.state_dict(), '/content/drive/MyDrive/snapshot/txtclassification.pt')
        best_val_loss = val_loss
```
